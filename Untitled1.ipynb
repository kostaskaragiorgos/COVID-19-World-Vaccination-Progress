```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vaccinations.csv')
df.head()
```

```python
indexlist = ["Asia", "Europe", "Africa", "Middle East", "World", "Upper middle income", "High income","North America", "Lower middle income"]
for i in indexlist:
    df[df['location']==str(i)].plot(figsize =(15,10),x='date',y=['total_vaccinations','people_vaccinated', 'people_fully_vaccinated' ], title="Vaccinations of "+str(i))
```
